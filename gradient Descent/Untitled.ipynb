```python
import numpy as np
```

```python
def cost(m,c,data):
    cost=0
    n=len(data)
    for i in range(n):
        y=data[i][1]
        x=data[i][0]
        cost+=(1/n)*((y-m*x-c)**2)
    return cost
```

```python
def cal(data,learning_curve,m_slope,c_inter):
    new_slope=0
    new_c=0
    for i in range(len(data)):
        new_slope+=(-2/len(data))*(data[i][1]-m_slope*data[i][0]-c_inter)*data[i][0]
        new_c+=(-2/len(data))*(data[i][1]-m_slope*data[i][0]-c_inter)
    new_m = m_slope - learning_curve*new_slope
    new_cc = c_inter - learning_curve*new_c
    return new_m, new_cc
```

```python
def gr(data,learning_curve,iteration):
    m_slope=0
    c_inter=0
    for i in range(iteration):
        m_slope,c_inter=cal(data,learning_curve,m_slope,c_inter)
        print(i,"cost: ",cost(m_slope,c_inter,data))
    return m_slope,c_inter
```

```python
def run():
    data=np.loadtxt('/Users/apple/Desktop/machine learning/gradient Descent/data.csv',delimiter=',')
    learning_curve=0.0001
    iteration=1000
    m,c=gr(data,learning_curve,iteration)
    print(m,c)
```

```python
run()
```
